**Confusion Matrix**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data.csv')

X=df.iloc[:,:-1]
y=df['Target']

# Write your code below
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)

model=BaggingClassifier(random_state=42)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
labels = np.asarray(
    [
        ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.flatten().sum())]
        for item in cm.flatten()
    ]
    ).reshape(cm.shape)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=labels, fmt="")
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.title('Confusion Matrix')
plt.show()
```

**random_forest with min_sample_leaf**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

df = pd.read_csv('data.csv')

# Write your code below
features=df[['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin']]
target=features['origin']
features.drop(columns='origin',inplace=True,axis=1)
features_encoded=pd.get_dummies(features,columns=['model year'],dtype=float)
X_train,X_test,y_train,y_test=train_test_split(features_encoded,target,test_size=0.3,random_state=1)
rforest=RandomForestClassifier(random_state=1,min_samples_leaf=3)
rforest.fit(X_train,y_train)
y_pred_train=rforest.predict(X_train)
train_accuracy=accuracy_score(y_pred_train,y_train)
```

**Random Forest with Class weight ='balanced'**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

df = pd.read_csv('data.csv')

# Write your code below
features=df[['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin']]
target=features['origin']
features.drop(columns='origin',inplace=True,axis=1)
features_encoded=pd.get_dummies(features,columns=['model year'],dtype=float)
X_train,X_test,y_train,y_test=train_test_split(features_encoded,target,test_size=0.3,random_state=1)
rforest=RandomForestClassifier(random_state=1,class_weight='balanced')
rforest.fit(X_train,y_train)
y_pred_train=rforest.predict(X_train)
train_accuracy=accuracy_score(y_pred_train,y_train)
```
